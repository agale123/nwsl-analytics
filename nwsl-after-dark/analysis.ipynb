```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load in data
df = pd.read_csv('matches.csv')
df
```

```python
# Function to plot scatter
def plot_stats(stat):
    x = pd.to_datetime(df['Date']).dt.hour
    y = df['Home ' + stat] + df['Away ' + stat]
    plt.scatter(x, y)
    m, b = np.polyfit(x, y, 1)
    plt.plot(x, m*x + b)
    plt.show()
```

```python
plot_stats('Goals')
```

```python
plot_stats('Fouls')
```

```python
plot_stats('Yellow Cards')
```

```python
plot_stats('Red Cards')
```
