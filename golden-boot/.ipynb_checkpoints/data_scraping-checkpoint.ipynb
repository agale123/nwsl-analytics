```python
import requests
import json
```

```python
data = requests.get("https://d2nkt8hgeld8zj.cloudfront.net/services/nwsl.ashx/playerstats?players=true").json()["data"]
print(data)
```

```python
season_id = data["season_id"]
season_id
```

```python
league_stats = {}
for p in data["players"]:
    if p["goals"] >= 5:
        player = p["player"]
        name = (
            player["lastName"]
            if player["firstName"] == " "
            else player["firstName"] + " " + player["lastName"]
        )
        slug = player["slug"]
        print(slug)
        player_data = requests.get(
            "https://d2nkt8hgeld8zj.cloudfront.net/services/nwsl.ashx/players?slug="
            + slug
        ).json()
        player_stats = {}
        for game in player_data["data"][0]["game_log"][season_id]:
            player_stats[game["game_date"][:10]] = game["game_stats"]["goals"]
        league_stats[name] = player_stats
league_stats
```

```python
with open("golden_boot.json", "w") as f:
    json.dump(league_stats, f)
```
