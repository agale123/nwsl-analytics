```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
```

```python
df = pd.read_csv('data.csv', index_col=0)
df
```

```python
colors = {
    'NJ/NY Gotham FC': '#000000',
    'Portland Thorns': '#97262c',
    'Angel City FC': '#000000',
    'Chicago Red Stars': '#3cb5e4',
    'Washington Spirit': '#c20032',
    'San Diego Wave': '#011e40',
    'North Carolina Courage': '#00416b',
    'OL Reign': '#002f87',
    'Racing Louisville': '#c5b4e1',
    'Houston Dash': '#ff6a01',
    'Orlando Pride': '#60269e',
    'Kansas City Current': '#62cac9',
}
```

```python
week=15
extra=3
for i in range(1, extra + 1):
    df[str(week+i)] = df[str(week)]
```

```python
barlow = fm.FontProperties(fname='/home/agale/nwsl-rank/Barlow.otf')
plt.rcParams['figure.figsize'] = (16, 9)
plt.figure(dpi=1200)
fig, ax = plt.subplots()

for team in df.index:
    ax.plot(
        df.columns,
        df.loc[team],
        linewidth=20,
        color=colors[team],
        solid_capstyle='round'
    )
    
    ax.text(
        week+2, 
        df.at[team, str(week)], 
        team, 
        color='white', 
        fontsize=14, 
        fontproperties=barlow,
        fontweight='bold',
        va='center',
        ha='right'
    )

# Title
plt.title('Week 15 Standings', fontproperties=barlow, size=32)
#plt.title('Week 15 Standings', fontsize=24, fontfamily='Barlow')
    
# General
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.tick_params(bottom=False, left=False)
ax.set_axisbelow(True)

# y-axis
fig.gca().invert_yaxis()
ax.set_yticks([])
ax.yaxis.grid(False)
# a-axis
ax.set_xticks(range(0, week))
ax.xaxis.grid(True, color='lightgray')
```
