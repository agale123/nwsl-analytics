```python
import json
import csv
```

```python
def load_matches(path):
    f = open(path)
    data = json.load(f)
    flat_matches = []
    for m in data["data"]["matches"]:
        for item in m["events"]:
            flat_matches.append(
                {
                    "date": item["date"].split('T')[0],
                    "home": item["team"]["abbreviation"],
                    "away": item["opponent"]["abbreviation"],
                }
            )
    return flat_matches
```

```python
matches = load_matches("schedule.json")
matches
```

```python
cc_matches = load_matches("challenge_cup_schedule.json")
cc_matches
```

```python
merged = sorted(matches + cc_matches, key=lambda x: x['date'])
```

```python
with open('schedule.csv', 'w') as f:
    writer = csv.writer(f)

    writer.writerow(['date', 'home', 'away'])
    
    for row in merged:
        writer.writerow([row['date'], row['home'], row['away']])
```
